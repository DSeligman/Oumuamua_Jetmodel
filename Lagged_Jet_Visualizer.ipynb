```python
#Oumuamua Jet Model
import time
from numpy import *
import matplotlib.pyplot as plt
%matplotlib inline  
import numpy as np
from scipy.integrate import odeint
```

```python
Lyc= np.loadtxt('Ly.txt')
Lzc = np.loadtxt('Lz.txt')
```

```python
endtime = 30.;
bigN = 1000000;

#converge at 1000 steps per .5 days
#20,000 steps per 1 day still converged

t = np.linspace(0, endtime*(60.*60.*24.), bigN)
plt.figure(figsize=(10,3))
plt.plot(t/(60.*60*24),np.sqrt(Lyc**2+Lzc**2),'purple',label='C++ RK4')
plt.xlabel('Time [Days]',fontsize=15)
plt.ylabel(r'$L \, / \, m \, \, [cm^2/s]$',fontsize=18)
plt.title('100 Minute Lag',fontsize=18)
#plt.ylabel('Total Angular Momentum',fontsize=15)
#plt.xlim([0,1])
#plt.plot(np.sqrt(Lyc**2+Lzc**2),'r--',label='C++ RK4')
#plt.legend(loc='upper left')
```

```python
plt.figure(figsize=(10,3))
plt.plot(t/(60.*60*24),np.sqrt(Lyc**2+Lzc**2),'purple',label='|L|')
plt.xlim([29,30])
plt.xlabel('Time [Days]',fontsize=15)
plt.ylabel(r'$L \, / \, m \, \, [cm^2/s]$',fontsize=18) 
plt.title('100 Minute Lag',fontsize=18)
```

```python
#plot all three
f, axarr = plt.subplots(2, sharex=True, sharey=False,figsize=(10,8))
#f.suptitle('Sharing both axes')
f.subplots_adjust(hspace=0.1)

axarr[0].plot(t/(60.*60*24),Lyc,label='Ly',color='purple')
axarr[0].plot(t/(60.*60*24),Lzc,label='Lz',color='grey')
axarr[0].plot(t/(60.*60*24),np.sqrt(Lyc**2+Lzc**2),color='black',label='|L|')    
axarr[0].legend(loc='lower center',ncol=3,prop={'size': 13},fancybox=True, shadow=True)
axarr[0].set_ylabel(r'$L \, / \, m \, \, [cm^2/s]$',fontsize=18)       
axarr[0].set_title('Converged Solution 6000 second Lag')
axarr[0].set_xlabel('Time [Days]',fontsize=15)


axarr[1].plot(t/(60.*60*24),Lyc,label='Ly',color='purple')
axarr[1].plot(t/(60.*60*24),Lzc,label='Lz',color='grey')
axarr[1].plot(t/(60.*60*24),np.sqrt(Lyc**2+Lzc**2),color='black',label='|L|')    
axarr[1].legend(loc='lower center',ncol=3,prop={'size': 13},fancybox=True, shadow=True)
axarr[1].set_ylabel(r'$L \, / \, m \, \, [cm^2/s]$',fontsize=18)       
axarr[1].set_title('Converged Solution 6000 second Lag')
axarr[1].set_xlabel('Time [Days]',fontsize=15)
axarr[1].set_xlim([29,30])

for ax in axarr:
    ax.label_outer()
```
